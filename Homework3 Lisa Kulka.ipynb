```r
#Question 2 (Ch 12, # 10) 

#2a)

#Generation of data set with 20 observations in each of three classes
set.seed(36)
x <- matrix(rnorm(20 * 3 * 50, mean = 0, sd = 0.001), ncol = 50)

#mean shift
x[1:20, 2] <- 1
x[21:40, 1] <- 2
x[21:40, 2] <- 2
x[41:60, 1] <- 1
true.labels <- c(rep(1, 20), rep(2, 20), rep(3, 20))
```

```r
#2b)

#Performing PCA
pca <- prcomp(x)
summary(pca)
```

```r
pca$x[,1:2]

#Creating a plot 
plot(pca$x[, 1:2], col = 1:3, xlab = "Z1", ylab = "Z2", pch = 19)

#The output shows that there are three distinct classes, and therefore we can continue to part c.
```

```r
#2c)

# Performing K-means clustering with K = 3
km.out <- kmeans(x, 3, nstart = 20)
table(true.labels, km.out$cluster)
```

```r
#After applying K-means clustering with K=3 to the observations, it is clear that the obtained clusters align perfectly with the true class labels. 
#This suggests that the K-means algorithm successfully identifies distinct groups within the data that correspond to the known class labels.
```

```r
#2d)

#Performing K-means clustering with K = 3
km.out <- kmeans(x, 2, nstart = 20)
table(true.labels, km.out$cluster)
```

```r
#After applying K-means clustering with K=2 to the observations, it is apparent that all of the observations from one of the original clusters 
#are grouped together in one of the two clusters obtained from the K-means clustering.
```

```r
#2e) 

#Performing K-means clustering with K = 4
km.out <- kmeans(x, 4, nstart = 20)
table(true.labels, km.out$cluster)
```

```r
#Upon performing K-means clustering with K=4 on the observations, 
#the initial cluster is shown to be divided into two distinct clusters.
```

```r
#2f)

#Performing K-means clustering with K = 3 on the first two principal component score vectors
km <- kmeans(pca$x[, 1:2], 3, nstart = 20)
table(true.labels, km$cluster)
```

```r
#When applying K-means clustering with K=3 to the first two 
#principal component core vectors of the observations, 
#the resulting clusters from K-means perfectly align with the true class labels.
```

```r
#2g)

#Performing K-means clustering with K = 3 with scaling
km.out <- kmeans(scale(x), 3, nstart = 20)
table(true.labels, km.out$cluster)
```

```r
#When performing clustering on the scaled data, 
#comparatively poorer results are shown relative to the unscaled data. 
#This is likely due to the effect of scaling on the distances between observations,
#which can impact the clustering outcome.
```

```r
#Question #3 

#3a)

#Loading data per instructions
gene_data <- read.csv("Ch12Ex13.csv", header=F)
head(gene_data)
```

```r
#3b)

#	Computing pairwise correlation matrix 
cor_matrix <- cor(gene_data)

#	Converting correlation matrix to dissimilarity matrix 
diss_matrix <- as.dist(1 - cor_matrix)

#	Performing hierarchical clustering
hc <- hclust(diss_matrix, method = "complete")

#	Plotting the dendrogram plot
plot(hc)


#Computing the correlation matrix 
cor_matrix <- cor(gene_data)

#Converting correlation matrix to dissimilarity matrix 
dissimilarity <- as.dist(1 - cor_matrix)

#Performing hierarchical clustering using different linkage methods
methods <- c("single", "complete", "average", "ward.D", "ward.D2", "centroid")
dendrograms <- list()

for (method in methods) {
#Hierarchical clustering
hc <- hclust(dissimilarity, method = method)
#Dendrogram plot
    dendrograms[[method]] <- hc
}

#Displaying the dendrograms
par(mfrow = c(2, 3))	#Setting the layout for multiple plots
for (i in 1:length(methods)) {
plot(dendrograms[[i]], main = paste("Dendrogram (", methods[i], "linkage)"))
}
```

```r
#The genes do separate the samples into the two groups.  In addition,
#almost all cluster in two groups except for the average linkage (3 groups).
#In terms of the results being dependent on the type of linkage used, 
#this does appear to be the case to an extent, as linkage methods invariably 
#influence clustering outcomes to some degree, though it should be noted that
#this appears to be minimal given that only one clusters into three groups, and
#all others cluster into two groups despite there being multiple linkage types 
#implemented across the analysis.  
```

```r
#3c) 

#To determine which group differs across the diseased patients and healthy patients, 
#it is arguably best to look at the loading vectors outputted from PCA 
#to see which genes are used to describe the variance the most.

pr = prcomp(t(gene_data))
summary(pr)
```

```r
#Based on this output, by performing K-means clustering with 2 clusters, 
#the genes that exhibit distinct expression values can be identified.

k2 <- kmeans(gene_data, centers=2, nstart=20)
k2$cluster
```

```r
#Results per the K-Means clustering with 2 clusters indicate that there 
#is significant dissimilarity between genes 11-20 and genes 500-600 in terms of their 
#expression levels across the two groups.
#To further validate this finding, principal component analysis (PCA) was performed. 
#The first principal component accounts for approximately 20% of the total variation 
#in the data and effectively separates the two groups. 
#This separation is visually evident when coloring the data points according to 
#the clusters generated by K-Means. The correct grouping of the two groups by K-Means 
#further supports the observation that genes 11-20 and genes 500-600 exhibit notable 
#differences between the two groups.

#Graphically representing these results
pc <- prcomp(gene_data, center=TRUE, scale. = TRUE)
par(mfrow=c(2,2))
biplot(pc)
plot(pc$x[,1], pc$x[,2], col=k2$cluster, pch=16, xlab="PC 1", ylab="PC 2")
plot(summary(pc)$importance[2,], type='l', lwd=2, ylab="Proportion Variance Explained", ylim= c(0,1))
plot(summary(pc)$importance[3,], type='l', lwd=2, ylab="Cumulative Variance Explained", ylim= c(0,1))
```

```r
#Question #4

#4a) Installation of all necessary libraries/packages per instructions

library(igraphdata)
data(karate)
karate
g <- karate

#Creating noisy datasets, inclusive of randomly deleting 5% of edges
set.seed(123)
percentage <- 0.05
num_edges <- ecount(g)
deleted_edges <- sample(1:num_edges, floor(num_edges * percentage))
g_noisy <- delete.edges(g, deleted_edges)

#Performing MCMC for random graph model
hrg <- hrg.fit (g_noisy, hrg = NULL, start = FALSE, steps = 0)

#Predicting missing edges
cat("Deleted edges: ", deleted_edges, "\n")

cat("Predictions:\n")

#Producing predictions
predict_edges(g_noisy, hrg=hrg)$edges
```
