```python
#Question 1

#1a) 

#Installing needed packages/libraries per assignment instructions
library(gRbase)
library(bnlearn)
data(cad1)
heart_data <- cad1

#Implementing the Hill-Climbing algorithm
algorithm <- "hc" 
#Establishing BIC score for model selection
params <- list(score = "bic") 
#Bayesian Network structure learning
bn_structure <- hc(heart_data, algorithm = algorithm, params = params)

#Viewing the resulting learned network structure
print(bn_structure)
```

```python
#Plotting the learned network structure
pchc::bnplot(bn_structure)
```

```python
#Model fitting based on results
bn_fitted <- bn.fit(bn_structure, data=cad1)
```

```python
#1b)

#In the process of getting all the d-separations, this shows my going through the 
#whole graph and testing each of the two nodes/changing the separating node. 
nodes <- nodes(bn_structure)
dsep_set <- c()
i_max <- length(nodes) - 1
j_max <- length(nodes)
for (k in 1:j_max){
for (i in 1:i_max){
    for (j in (i+1):j_max){
        node_a <- nodes[i]
        node_b <- nodes[j]
        if (dsep(bn_fitted, node_a, node_b, nodes[k])){
        dsep_set <- rbind(dsep_set, c(node_a, node_b, nodes[k]))
        }
    }
}
}
print("All d-separations: (node a, node b, separating node")
```

```python
#2e)

#Installing needed packages/libraries per assignment instructions
library(cluster) 

#Performing k-means clustering (with range of values indicated and calculating silhouette scores
k_values <- 2:8 
silhouette_scores <- vector("numeric", length(k_values)) 
scores
for (k in k_values) {
    kmeans_result <- kmeans(wines, centers = k)
    silhouette_scores[k-1] <- silhouette(kmeans_result$cluster, dist(wines))
}

#Plotting the resutling silhouette scores against k
plot(k_values, silhouette_scores, type = "b", pch = 19, frame = FALSE,
xlab = "Number of Clusters (k)", ylab = "Silhouette Score", main = "Silhouette Analysis")
```

```python
#Determining the optimal k value
optimal_k <- k_values[which.max(silhouette_scores)]

#Viewing the results
cat("Optimal number of clusters (k):", optimal_k, "\n")
```

```python
#6b)


#Installing needed packages/libraries per assignment instructionsd required packages
install.packages("glasso")
install.packages("corpcor")
library(glasso)
library(corpcor)

#Loading the state data
data(state)
state_data <- data.frame(state.x77)

#Highlighting required columns
state_data <- state_data[, c("Population", "Income", "Illiteracy", "Life.Exp", "Murder", "HS.Grad", "Frost", "Area")]

#Scaling the data
state_data_scaled <- scale(state_data)

#Calculating the covariance matrix of the normalized data
cov_matrix <- cov(state_data_scaled)

#Defining a range of penalty values
penalty_values <- c(0.01, 0.05, 0.1, 0.2)

#Applying Graphical Lasso for each penalty value
lasso_results <- lapply(penalty_values, function(penalty) {
  glasso(cov_matrix, rho = penalty)
})

#Extracting and analyzing precision matrices
precision_matrices <- lapply(lasso_results, function(result) result$wi)

#Viewing the precision matrices for different penalty values
for (i in 1:length(penalty_values)) {
  cat("Precision Matrix for Penalty =", penalty_values[i], "\n")
  print(precision_matrices[[i]])
  cat("\n")
}

#Visualizing the precision matrices
par(mfrow = c(2, 2))
for (i in 1:length(penalty_values)) {
  image(precision_matrices[[i]], main = paste("Precision Matrix (Penalty =", penalty_values[i], ")"), xlab = "", ylab = "")
}

```

```python
#In terms of findings for each of the penalties, varying insights 
#were found; for Penalty = 0.01, the precision matrix shows many 
#non-zero entries, and significant relationships are clear between 
#Income and HS Grad, and Population and Area, suggesting robust 
#economic and demographic connections.  For Penalty = 0.05, 
#the precision matrix is comparatively more sparse, with key 
#dependencies inclusive of the negative associations between 
#Illiteracy and Life Exp and Murder and Life Exp; this serves 
#to highlight important socioeconomic and health relationships.  
#For Penalty = 0.1, the precision matrix becomes even more sparse, 
#creating an emphasis on stronger dependencies (i.e. the relationship 
#between Income and HS Grad, as well as a notable link between 
#Frost and Life Exp.) This higher penalty value filters out more 
#noise and emphasizes the most salient relationships.  Finally, 
#the Penalty = 0.2is the most sparse, with only the dependencies 
#relative to Population and Area as well as Income and HS Grad 
#showing; this means that the most statistically important demographic 
#and economic connections remain at the point of the most rigorous 
#testing, and should be the most emphasized. In terms of comparison 
#with 5a), the SOM clustering did show overall similarities in terms 
#of relationships of note (i.e. significant relationships between 
#population, income, education, life expectancy, murder rates, frost days, 
#and area, but the testing was not inclusive of consideration of 
#conditional dependencies and the results are therefore less statistically 
#robust.  The grouping of states was based on combined effects of variables, 
#whereas this testing was based direct conditional dependencies between 
#variables, resulting in the strongest representation of the most significant 
#relationships.
```
